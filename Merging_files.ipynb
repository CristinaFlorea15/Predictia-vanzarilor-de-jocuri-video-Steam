```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.read_csv("info_base_games.csv", low_memory=False)
df2 = pd.read_csv("dlcs.csv")
df3 = pd.read_csv("demos.csv")
df4 = pd.read_csv("gamalytic_steam_games.csv")
df5 = pd.read_csv("scraped_data.csv")
```

```python
df1['has_dlc'] = df1['appid'].isin(df2['base_appid']).astype(int)
```

```python
df1['has_demo'] = df1['appid'].isin(df3['full_game_appid']).astype(int)
```

```python
merged_df = pd.merge(
    df1,
    df4,
    how='inner',
    left_on='appid',
    right_on='steamId',
    suffixes=('_info', '_gamalytic')
)

merged_df.to_csv("merged_df1_df4.csv", index=False)
```

```python
final_df = pd.merge(
    merged_df,
    df5,
    how='inner',
    left_on='steamId',
    right_on='steamId',
    suffixes=('', '_scraped')
)

if 'steamId' in final_df.columns and 'appid' in final_df.columns:
    final_df.drop(columns=['steamId'], inplace=True)

print(f"Total rows: {final_df.shape[0]}")
print(f"Total columns: {final_df.shape[1]}")
print("\nColumns:\n", final_df.columns.tolist())
```
